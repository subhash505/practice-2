```python
l1 = [(0, 1), (2, 3), (4, 5), (6, 6), (7, 7), (8, 9)]
l2 =[(0, 4), (5, 14), (15, 23), (24, 24), (25, 34), (35, 45), (46, 48)]

for f1, f2 in l2:
    for e1, e2 in l1 :
        df1.loc[f2+1] = df.loc[e1:e2].values
```

```python
grouped_df1 = df1.groupby('Custodian Acct')
grouped_df2 = df.groupby('Custodian Acct')

merged_dfs = []
for custodian, group_df1 in grouped_df1:
    print(f"custodian {custodian} and group_f1 {group_df1}")
    if custodian in grouped_df2.groups:
        group_df2 = grouped_df2.get_group(custodian)
        merged_df = pd.concat([group_df1, group_df2], ignore_index=True)
    else:
        merged_df = group_df1.copy()
    merged_dfs.append(merged_df)
result_df = pd.concat(merged_dfs, ignore_index=True)
result_df.to_csv("merge_file.csv")
```
